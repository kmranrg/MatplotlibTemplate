Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
```

```python
plt.scatter(day,no)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
plt.scatter(day,no,color="r")
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
plt.scatter(day,no,color=colors)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,color=colors,s=sizes)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,color=colors,s=sizes,alpha=0.4)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,color=colors,s=sizes,alpha=1,marker='+')
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,color=colors,s=sizes,alpha=1,marker="*",edgecolor="g")
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=['r','y','g','b','r','g','r']
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,color=colors,s=sizes,alpha=1,marker="*",edgecolor="g",linewidth=3)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=[10,49,30,29,56,20,30]
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
colors=[10,49,30,29,56,20,30]
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no,c=colors,s=sizes,cmap='BrBG')
cb=plt.colorbar()
cb.set_label('ColorBar')
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no_1=[2,3,1,4,5,3,6]
no_2=[3,2,4,5,1,6,2]

colors=[10,49,30,29,56,20,30]
sizes=[100,200,1,45,999,12,9]
plt.scatter(day,no_1,c=colors,s=sizes,cmap='BrBG')
plt.scatter(day,no_2,c='r',s=sizes)
cb=plt.colorbar()
cb.set_label('ColorBar')
plt.title('Scatter Plot',fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
