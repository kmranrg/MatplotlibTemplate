Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['python', 'c', 'c++', 'java']
y = [85, 40, 60, 82]
z = [20, 30, 40, 60]
```

```python
plt.xlabel('language',fontsize=15)
plt.ylabel('number',fontsize=15)
plt.title('Bar Graph',fontsize=15)
plt.bar(x,y,width=0.4,color='g',alpha=0.4,label='popularity_2019')
plt.bar(x,z,width=0.4,color='r',alpha=1,label='popularity_2020')
plt.legend()
plt.show()
```

```python
width = 0.2
bar1 = np.arange(len(x))
bar2 = [j+width for j in bar1]

plt.xlabel('language',fontsize=15)
plt.ylabel('number',fontsize=15)
plt.title('Bar Graph',fontsize=15)

plt.bar(bar1,y,width=width,color='g',alpha=0.4,label='popularity_2019')
plt.bar(bar2,z,width=width,color='r',alpha=1,label='popularity_2020')

plt.xticks(bar1+width/2,x)
plt.legend()
plt.show()
```

```python
width = 0.2
bar1 = np.arange(len(x))
bar2 = [j+width for j in bar1]

plt.xlabel('language',fontsize=15)
plt.ylabel('number',fontsize=15)
plt.title('Bar Graph',fontsize=15)

plt.bar(bar1,y,width=width,color='g',alpha=0.4,label='popularity_2019')
plt.bar(bar2,z,width=width,color='r',alpha=1,label='popularity_2020')

plt.xticks(bar1+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
plt.xlabel('number',fontsize=15)
plt.ylabel('language',fontsize=15)
plt.title('Bar Graph',fontsize=15)
plt.barh(x,y,color='g',alpha=0.4,label='popularity_2019')
plt.barh(x,z,color='r',alpha=1,label='popularity_2020')
plt.legend()
plt.show()
```
